Which games are possible with 12 red cubes, 13 green cubes, and 14 blue cubes?

Maybe create 3 new columns, red_sum, blue_sum, and green_sum?

```python
def red_max(row):
    red_values = [int(s) for value in row if 'red' in str(value) for s in str(value).split() if s.isdigit()]
    return np.max(red_values) if red_values else 0

def blue_max(row):
    blue_values = [int(s) for value in row if 'blue' in str(value) for s in str(value).split() if s.isdigit()]
    return np.max(blue_values) if blue_values else 0

def green_max(row):
    green_values = [int(s) for value in row if 'green' in str(value) for s in str(value).split() if s.isdigit()]
    return np.max(green_values) if green_values else 0
```

```python
# Create new columns for the sum of colors
df['red_max'] = df.apply(red_max, axis = 1)
df['blue_max'] = df.apply(blue_max, axis = 1)
df['green_max'] = df.apply(green_max, axis = 1)

# Create conditions for a filtered dataframe
red_condition = df['red_max'] <= 12
blue_condition = df['blue_max'] <= 14
green_condition = df['green_max'] <= 13

#Locate rows that satisfy all conditions and create a filtered df from it

#filtered_df = df[red_condition & blue_condition & green_condition]

#Sum the values of the game IDs
#filtered_df
df
```
